```python
import networkx as nx
G=nx.Graph()
```

```python
G.add_edges_from([(1,2),(1,3),(1,4),(2,3),(3,4),(4,5),(4,6),(5,6),(5,7),(5,8),(6,7),(6,8),(7,8),(7,9)])
```

```python
nx.draw(G,with_labels=True, arrowstyle="->")
```

## Degree Centrainty

```python
Dc = nx.degree_centrality(G)
Dc
```

```python
sorted(Dc, key=Dc.get, reverse=True)
```

## Closeness Centrality

```python
Cc = nx.closeness_centrality(G)
Cc
```

```python
sorted(Cc, key=Cc.get, reverse=True)
```

## Betweenness Centrality

```python
Bc = nx.betweenness_centrality(G)
Bc
```

```python
sorted(Bc, key=Bc.get, reverse=True)
```

## Eigenvector Centrality

```python
Ec = nx.eigenvector_centrality(G)
Ec
```

```python
sorted(Ec, key=Ec.get, reverse=True)
```
